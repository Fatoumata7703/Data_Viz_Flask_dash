```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
import sys
import os
import warnings
warnings.filterwarnings('ignore')

if 'banque_selectionnee' not in dir():
    banque_selectionnee = 'BNDE'
if os.getcwd() not in sys.path:
    sys.path.insert(0, os.getcwd())

# Navy & Gold theme
NAVY = '#0f172a'
NAVY_DARK = '#0f172a'
BLUE = '#0891b2'
GOLD = '#f59e0b'
GOLD_LIGHT = '#fbbf24'
TEXT = '#1e293b'
MUTED = '#64748b'
BG = '#f1f5f9'
BORDER = '#e2e8f0'
RED = '#ef4444'
GREEN = '#10b981'

plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Inter', 'Segoe UI', 'Arial'],
    'font.size': 11,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'axes.facecolor': BG,
    'figure.facecolor': '#ffffff',
    'axes.edgecolor': BORDER,
    'axes.grid': True,
    'grid.alpha': 0.3,
    'grid.linewidth': 0.5,
    'grid.linestyle': '--',
    'text.color': TEXT,
    'axes.labelcolor': TEXT,
    'xtick.color': MUTED,
    'ytick.color': MUTED,
})
```

```python
try:
    from data_loader import get_data
    df = get_data()
except Exception:
    df = None
if df is None or len(df) == 0:
    df = pd.read_excel('data/BASE_SENEGAL2.xlsx')

df_full = df.copy()

banque_col = 'Sigle'
annee_col = 'ANNEE'
groupe_col = 'Goupe_Bancaire' if 'Goupe_Bancaire' in df_full.columns else 'Groupe_Bancaire'

df_full[annee_col] = pd.to_numeric(df_full[annee_col], errors='coerce')
annees = sorted(df_full[annee_col].dropna().astype(int).unique())
if 'annee_rapport' in dir() and annee_rapport is not None:
    annee_cible = int(annee_rapport)
else:
    annee_cible = int(annees[-1]) if annees else 2020
annee_debut = int(annees[0]) if annees else 2015
annee_fin = annee_cible

banque_rapport = banque_selectionnee if 'banque_selectionnee' in dir() and banque_selectionnee else 'BNDE'

num_cols = ['BILAN', 'EMPLOI', 'RESSOURCES', 'FONDS_PROPRES',
            'produit_net_bancaire', 'resultat_net', 'effectif', 'agence',
            "charges_generales_d'exploitation", 'cout_du_risque',
            'resultat_brut_d_exploitation', 'resultat_d_exploitation',
            'resultat_avant_impot', 'COMPTE']
for c in num_cols:
    if c in df_full.columns:
        df_full[c] = pd.to_numeric(df_full[c], errors='coerce')
```

```python
from IPython.display import display, HTML

df_year = df_full[df_full[annee_col] == annee_cible].copy()
nb_banques = df_year[banque_col].nunique()
total_bilan = df_year['BILAN'].sum() if 'BILAN' in df_year.columns else 0
total_pnb = df_year['produit_net_bancaire'].sum() if 'produit_net_bancaire' in df_year.columns else 0
total_rn = df_year['resultat_net'].sum() if 'resultat_net' in df_year.columns else 0
total_fp = df_year['FONDS_PROPRES'].sum() if 'FONDS_PROPRES' in df_year.columns else 0
total_emploi = df_year['EMPLOI'].sum() if 'EMPLOI' in df_year.columns else 0
total_ress = df_year['RESSOURCES'].sum() if 'RESSOURCES' in df_year.columns else 0
total_eff = int(df_year['effectif'].sum()) if 'effectif' in df_year.columns else 0
total_ag = int(df_year['agence'].sum()) if 'agence' in df_year.columns else 0

def fmt(v):
    if abs(v) >= 1e9: return f"{v/1e9:,.1f} Mds"
    if abs(v) >= 1e6: return f"{v/1e6:,.0f} M"
    if abs(v) >= 1e3: return f"{v/1e3:,.0f} k"
    return f"{v:,.0f}"

kpis = [
    ('Banques actives', str(nb_banques), ''),
    ('Total Bilan', fmt(total_bilan), 'FCFA'),
    ('Produit Net Bancaire', fmt(total_pnb), 'FCFA'),
    ('Résultat Net', fmt(total_rn), 'FCFA'),
    ('Fonds Propres', fmt(total_fp), 'FCFA'),
    ('Emplois', fmt(total_emploi), 'FCFA'),
    ('Ressources', fmt(total_ress), 'FCFA'),
    ('Effectifs', f"{total_eff:,}", 'collaborateurs'),
]

cards_html = ''
for label, value, unit in kpis:
    cards_html += f'''
    <div style="flex: 1; min-width: 180px; background: #fff; border-radius: 12px;
                padding: 18px 16px; box-shadow: 0 2px 12px rgba(26,41,64,0.06);
                border: 1px solid #e2e8f0; text-align: center;">
      <div style="font-size: 0.72rem; font-weight: 600; color: #64748b;
                  text-transform: uppercase; letter-spacing: 0.05em; margin-bottom: 6px;">{label}</div>
      <div style="font-size: 1.4rem; font-weight: 800; color: #0f172a;">{value}</div>
      <div style="font-size: 0.7rem; color: #94a3b8; margin-top: 2px;">{unit}</div>
    </div>'''

display(HTML(f'''
<div style="font-family: 'Inter', 'Segoe UI', sans-serif; margin: 12px 0;">
  <div style="background: linear-gradient(135deg, #0f172a, #0891b2); color: #fff;
              padding: 14px 22px; border-radius: 12px 12px 0 0; display: flex;
              align-items: center; gap: 10px;">
    <span style="font-size: 1.1rem;">📊</span>
    <span style="font-weight: 700; font-size: 0.95rem;">Indicateurs clés — Secteur bancaire sénégalais ({annee_cible})</span>
  </div>
  <div style="display: flex; flex-wrap: wrap; gap: 12px; padding: 16px;
              background: #f1f5f9; border-radius: 0 0 12px 12px;
              border: 1px solid #e2e8f0; border-top: none;">
    {cards_html}
  </div>
</div>
'''))
```

```python
# Évolution du total bilan sectoriel
if 'BILAN' in df_full.columns:
    evo = df_full.groupby(annee_col)['BILAN'].sum().reset_index().sort_values(annee_col)
    evo = evo[evo[annee_col].between(annee_debut, annee_fin)]
    evo['croissance'] = evo['BILAN'].pct_change() * 100

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle(f'Dynamique du secteur bancaire sénégalais ({annee_debut}–{annee_fin})',
                 fontsize=15, fontweight='bold', color=NAVY, y=1.02)

    bars = ax1.bar(evo[annee_col].astype(int), evo['BILAN'] / 1e6,
                   color=BLUE, alpha=0.85, edgecolor='white', linewidth=1.2, width=0.65)
    for bar, val in zip(bars, evo['BILAN']):
        ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + evo['BILAN'].max()/1e6*0.01,
                 fmt(val), ha='center', va='bottom', fontsize=9, fontweight='600', color=NAVY)
    ax1.set_title('Total Bilan Sectoriel', fontweight='bold', color=NAVY, pad=10)
    ax1.set_ylabel('Milliards FCFA', fontweight='600')
    ax1.set_xlabel('')
    for sp in ['top', 'right']: ax1.spines[sp].set_visible(False)

    valid_crois = evo.dropna(subset=['croissance'])
    colors_bar = [GREEN if v >= 0 else RED for v in valid_crois['croissance']]
    ax2.bar(valid_crois[annee_col].astype(int), valid_crois['croissance'],
            color=colors_bar, alpha=0.85, edgecolor='white', linewidth=1.2, width=0.65)
    for x, v in zip(valid_crois[annee_col].astype(int), valid_crois['croissance']):
        ax2.text(x, v + (0.3 if v >= 0 else -0.5), f'{v:.1f}%',
                 ha='center', va='bottom' if v >= 0 else 'top', fontsize=9, fontweight='600', color=NAVY)
    ax2.axhline(y=0, color=BORDER, linewidth=1)
    ax2.set_title('Croissance annuelle (%)', fontweight='bold', color=NAVY, pad=10)
    ax2.set_ylabel('Taux de croissance (%)', fontweight='600')
    for sp in ['top', 'right']: ax2.spines[sp].set_visible(False)

    plt.tight_layout()
    plt.show()
```

```python
if 'BILAN' in df_full.columns:
    df_bar = df_year.groupby(banque_col)['BILAN'].sum().reset_index()
    df_bar = df_bar.sort_values('BILAN', ascending=True)

    colors = [RED if str(b).strip().upper() == str(banque_rapport).strip().upper() else BLUE
              for b in df_bar[banque_col]]

    fig, ax = plt.subplots(figsize=(14, max(8, len(df_bar) * 0.5)))
    bars = ax.barh(df_bar[banque_col], df_bar['BILAN'],
                   color=colors, alpha=0.88, edgecolor='white', linewidth=1)

    max_val = df_bar['BILAN'].max()
    for bar, val in zip(bars, df_bar['BILAN']):
        ax.text(val + max_val * 0.005, bar.get_y() + bar.get_height()/2,
                fmt(val), va='center', ha='left', fontsize=9, color=TEXT, fontweight='500')

    ax.set_xlabel('Total Bilan (Millions FCFA)', fontsize=12, fontweight='600', color=TEXT)
    ax.set_title(f'CLASSEMENT DES BANQUES PAR TOTAL BILAN — {annee_cible}',
                 fontsize=14, fontweight='bold', pad=14, color=NAVY)

    ax.text(0.98, 0.98, f'▶ {banque_rapport}', transform=ax.transAxes,
            fontsize=10, ha='right', va='top', color=RED, fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.4', facecolor='#fff', edgecolor=RED, linewidth=1.5))

    ax.text(0.02, 0.02, 'Source: Commission Bancaire, Analyses Performances Group',
            transform=ax.transAxes, fontsize=8, style='italic', color=MUTED,
            bbox=dict(boxstyle='round', facecolor='#fff', edgecolor=BORDER, alpha=0.9))

    for sp in ['top', 'right']: ax.spines[sp].set_visible(False)
    ax.spines['left'].set_color(BORDER)
    ax.spines['bottom'].set_color(BORDER)
    ax.set_xlim(left=0, right=max_val * 1.15)
    plt.tight_layout()
    plt.show()
```

```python
# Parts de marché — Top 10
if 'BILAN' in df_full.columns:
    df_parts = df_year.groupby(banque_col)['BILAN'].sum().reset_index()
    total = df_parts['BILAN'].sum()
    df_parts['Part (%)'] = (df_parts['BILAN'] / total * 100).round(2)
    df_parts = df_parts.sort_values('BILAN', ascending=False)
    top10 = df_parts.head(10).copy()
    autres = pd.DataFrame({banque_col: ['Autres'], 'BILAN': [df_parts.iloc[10:]['BILAN'].sum()],
                           'Part (%)': [df_parts.iloc[10:]['Part (%)'].sum()]})
    pie_data = pd.concat([top10, autres], ignore_index=True)

    palette = [NAVY, BLUE, '#4a7ab5', GOLD, GOLD_LIGHT, GREEN,
               '#0ea5e9', '#8b5cf6', '#f59e0b', '#ec4899', '#94a3b8']

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), gridspec_kw={'width_ratios': [1, 1.2]})
    fig.suptitle(f'Concentration du marché bancaire — {annee_cible}',
                 fontsize=14, fontweight='bold', color=NAVY, y=1.02)

    wedges, texts, autotexts = ax1.pie(
        pie_data['BILAN'], labels=pie_data[banque_col],
        autopct='%1.1f%%', colors=palette[:len(pie_data)],
        startangle=90, pctdistance=0.8,
        wedgeprops=dict(width=0.45, edgecolor='white', linewidth=2))
    for t in texts: t.set_fontsize(8)
    for t in autotexts: t.set_fontsize(7); t.set_fontweight('bold')
    ax1.set_title('Répartition par total bilan', fontweight='bold', color=NAVY, pad=10)

    hhi = ((df_parts['Part (%)'])**2).sum()
    top3_share = df_parts.head(3)['Part (%)'].sum()
    top5_share = df_parts.head(5)['Part (%)'].sum()

    concentration_data = {
        'Indicateur': ['HHI (Herfindahl)', 'Part Top 3', 'Part Top 5', 'Part Top 10', 'Nb banques'],
        'Valeur': [f'{hhi:.0f}', f'{top3_share:.1f}%', f'{top5_share:.1f}%',
                   f'{top10["Part (%)"].sum():.1f}%', str(nb_banques)],
        'Interprétation': [
            'Concentré' if hhi > 2500 else ('Modéré' if hhi > 1500 else 'Compétitif'),
            'Forte dominance' if top3_share > 50 else 'Répartition équilibrée',
            'Oligopole' if top5_share > 70 else 'Marché diversifié',
            '', ''
        ]
    }
    df_conc = pd.DataFrame(concentration_data)

    ax2.axis('off')
    table = ax2.table(cellText=df_conc.values, colLabels=df_conc.columns,
                      loc='center', cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(10)
    table.scale(1, 1.8)
    for (i, j), cell in table.get_celld().items():
        if i == 0:
            cell.set_facecolor(NAVY)
            cell.set_text_props(color='white', fontweight='bold')
        else:
            cell.set_facecolor('#f1f5f9' if i % 2 == 0 else '#fff')
        cell.set_edgecolor(BORDER)
    ax2.set_title('Indicateurs de concentration', fontweight='bold', color=NAVY, pad=10)

    plt.tight_layout()
    plt.show()
```

```python
df_bk = df_full[df_full[banque_col].str.strip().str.upper() == str(banque_rapport).strip().upper()].copy()
df_bk_year = df_bk[df_bk[annee_col] == annee_cible]

bk_bilan = df_bk_year['BILAN'].sum() if 'BILAN' in df_bk_year.columns else 0
bk_pnb = df_bk_year['produit_net_bancaire'].sum() if 'produit_net_bancaire' in df_bk_year.columns else 0
bk_rn = df_bk_year['resultat_net'].sum() if 'resultat_net' in df_bk_year.columns else 0
bk_fp = df_bk_year['FONDS_PROPRES'].sum() if 'FONDS_PROPRES' in df_bk_year.columns else 0
bk_eff = int(df_bk_year['effectif'].sum()) if 'effectif' in df_bk_year.columns else 0
bk_ag = int(df_bk_year['agence'].sum()) if 'agence' in df_bk_year.columns else 0

part_bilan = (bk_bilan / total_bilan * 100) if total_bilan > 0 else 0
roe = (bk_rn / bk_fp * 100) if bk_fp > 0 else 0

bk_charges = df_bk_year["charges_generales_d'exploitation"].sum() if "charges_generales_d'exploitation" in df_bk_year.columns else 0
coef_exploit = (bk_charges / bk_pnb * 100) if bk_pnb > 0 else 0

rank_bilan = df_year.groupby(banque_col)['BILAN'].sum().rank(ascending=False, method='min')
bk_rank = int(rank_bilan.get(banque_rapport, 0))

fiche_kpis = [
    ('Total Bilan', fmt(bk_bilan), 'FCFA'),
    ('Rang Bilan', f'{bk_rank}e / {nb_banques}', ''),
    ('Part de marché', f'{part_bilan:.1f}%', 'du secteur'),
    ('PNB', fmt(bk_pnb), 'FCFA'),
    ('Résultat Net', fmt(bk_rn), 'FCFA'),
    ('ROE', f'{roe:.1f}%', ''),
    ('Coef. Exploitation', f'{coef_exploit:.1f}%', ''),
    ('Effectifs', f'{bk_eff:,}', f'{bk_ag} agences'),
]

cards_html = ''
for label, value, unit in fiche_kpis:
    cards_html += f'''
    <div style="flex: 1; min-width: 160px; background: #fff; border-radius: 12px;
                padding: 16px 14px; box-shadow: 0 2px 12px rgba(26,41,64,0.06);
                border: 1px solid #e2e8f0; text-align: center;">
      <div style="font-size: 0.7rem; font-weight: 600; color: #64748b;
                  text-transform: uppercase; letter-spacing: 0.04em; margin-bottom: 5px;">{label}</div>
      <div style="font-size: 1.25rem; font-weight: 800; color: #0f172a;">{value}</div>
      <div style="font-size: 0.68rem; color: #94a3b8; margin-top: 2px;">{unit}</div>
    </div>'''

groupe_bk = df_bk_year[groupe_col].iloc[0] if len(df_bk_year) > 0 and groupe_col in df_bk_year.columns else 'N/A'

display(HTML(f'''
<div style="font-family: 'Inter', 'Segoe UI', sans-serif; margin: 12px 0;">
  <div style="background: linear-gradient(135deg, #0f172a, #0891b2); color: #fff;
              padding: 14px 22px; border-radius: 12px 12px 0 0; display: flex;
              justify-content: space-between; align-items: center;">
    <div>
      <span style="font-weight: 800; font-size: 1.1rem;">🏦 {banque_rapport}</span>
      <span style="font-size: 0.8rem; opacity: 0.8; margin-left: 12px;">{groupe_bk}</span>
    </div>
    <span style="font-size: 0.8rem; background: rgba(212,168,67,0.2); color: #fbbf24;
                 padding: 3px 12px; border-radius: 20px; font-weight: 600;">{annee_cible}</span>
  </div>
  <div style="display: flex; flex-wrap: wrap; gap: 10px; padding: 14px;
              background: #f1f5f9; border-radius: 0 0 12px 12px;
              border: 1px solid #e2e8f0; border-top: none;">
    {cards_html}
  </div>
</div>
'''))
```

```python
has_pnb = 'produit_net_bancaire' in df_year.columns
has_rn = 'resultat_net' in df_year.columns
has_fp = 'FONDS_PROPRES' in df_year.columns
has_charges = "charges_generales_d'exploitation" in df_year.columns

if has_pnb and has_rn:
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle(f'Analyse de rentabilité — Secteur bancaire ({annee_cible})',
                 fontsize=15, fontweight='bold', color=NAVY, y=1.01)

    # PNB Top 10
    ax = axes[0, 0]
    df_pnb = df_year.groupby(banque_col)['produit_net_bancaire'].sum().reset_index()
    df_pnb = df_pnb.sort_values('produit_net_bancaire', ascending=True).tail(10)
    c = [RED if str(b).strip().upper() == str(banque_rapport).strip().upper() else BLUE
         for b in df_pnb[banque_col]]
    ax.barh(df_pnb[banque_col], df_pnb['produit_net_bancaire'], color=c, alpha=0.88, edgecolor='white')
    ax.set_title('Top 10 — Produit Net Bancaire', fontweight='bold', color=NAVY, pad=10)
    ax.set_xlabel('PNB (Millions FCFA)')
    for sp in ['top', 'right']: ax.spines[sp].set_visible(False)

    # Résultat Net Top 10
    ax = axes[0, 1]
    df_rn = df_year.groupby(banque_col)['resultat_net'].sum().reset_index()
    df_rn = df_rn.sort_values('resultat_net', ascending=True).tail(10)
    c = [RED if str(b).strip().upper() == str(banque_rapport).strip().upper() else
         GREEN if v >= 0 else '#ef4444'
         for b, v in zip(df_rn[banque_col], df_rn['resultat_net'])]
    ax.barh(df_rn[banque_col], df_rn['resultat_net'], color=c, alpha=0.88, edgecolor='white')
    ax.set_title('Top 10 — Résultat Net', fontweight='bold', color=NAVY, pad=10)
    ax.set_xlabel('Résultat Net (Millions FCFA)')
    for sp in ['top', 'right']: ax.spines[sp].set_visible(False)

    # ROE
    ax = axes[1, 0]
    if has_fp:
        df_roe = df_year.groupby(banque_col).agg({'resultat_net': 'sum', 'FONDS_PROPRES': 'sum'}).reset_index()
        df_roe = df_roe[df_roe['FONDS_PROPRES'] > 0]
        df_roe['ROE'] = (df_roe['resultat_net'] / df_roe['FONDS_PROPRES'] * 100).round(1)
        df_roe = df_roe.sort_values('ROE', ascending=True).tail(12)
        c = [RED if str(b).strip().upper() == str(banque_rapport).strip().upper() else
             GOLD if v >= 0 else '#ef4444'
             for b, v in zip(df_roe[banque_col], df_roe['ROE'])]
        ax.barh(df_roe[banque_col], df_roe['ROE'], color=c, alpha=0.88, edgecolor='white')
        ax.axvline(x=0, color=BORDER, linewidth=1)
    ax.set_title('Rendement des fonds propres (ROE)', fontweight='bold', color=NAVY, pad=10)
    ax.set_xlabel('ROE (%)')
    for sp in ['top', 'right']: ax.spines[sp].set_visible(False)

    # Coefficient d'exploitation
    ax = axes[1, 1]
    if has_charges:
        df_coef = df_year.groupby(banque_col).agg(
            {'produit_net_bancaire': 'sum', "charges_generales_d'exploitation": 'sum'}).reset_index()
        df_coef = df_coef[df_coef['produit_net_bancaire'] > 0]
        df_coef['Coef'] = (df_coef["charges_generales_d'exploitation"] / df_coef['produit_net_bancaire'] * 100).round(1)
        df_coef = df_coef.sort_values('Coef', ascending=True).tail(12)
        c = [RED if str(b).strip().upper() == str(banque_rapport).strip().upper() else
             '#0ea5e9' for b in df_coef[banque_col]]
        ax.barh(df_coef[banque_col], df_coef['Coef'], color=c, alpha=0.88, edgecolor='white')
        ax.axvline(x=60, color=GOLD, linewidth=1.5, linestyle='--', alpha=0.6)
        ax.text(61, len(df_coef)-1, 'Seuil 60%', fontsize=8, color=GOLD, fontweight='600')
    ax.set_title("Coefficient d'exploitation", fontweight='bold', color=NAVY, pad=10)
    ax.set_xlabel('Coef. exploitation (%)')
    for sp in ['top', 'right']: ax.spines[sp].set_visible(False)

    plt.tight_layout()
    plt.show()
```

```python
agg_cols = {}
for c in ['BILAN', 'EMPLOI', 'RESSOURCES', 'FONDS_PROPRES',
           'produit_net_bancaire', 'resultat_net', 'effectif', 'agence']:
    if c in df_year.columns:
        agg_cols[c] = 'sum'

if agg_cols:
    df_comp = df_year.groupby(banque_col).agg(agg_cols).reset_index()

    if 'BILAN' in df_comp.columns:
        total_b = df_comp['BILAN'].sum()
        df_comp['Part Marché (%)'] = (df_comp['BILAN'] / total_b * 100).round(2)

    if 'resultat_net' in df_comp.columns and 'FONDS_PROPRES' in df_comp.columns:
        df_comp['ROE (%)'] = np.where(
            df_comp['FONDS_PROPRES'] > 0,
            (df_comp['resultat_net'] / df_comp['FONDS_PROPRES'] * 100).round(1), 0)

    if 'produit_net_bancaire' in df_comp.columns and "charges_generales_d'exploitation" in df_year.columns:
        charges = df_year.groupby(banque_col)["charges_generales_d'exploitation"].sum()
        df_comp = df_comp.set_index(banque_col)
        df_comp['Coef. Exploit. (%)'] = np.where(
            df_comp['produit_net_bancaire'] > 0,
            (charges / df_comp['produit_net_bancaire'] * 100).round(1), 0)
        df_comp = df_comp.reset_index()

    df_comp = df_comp.sort_values('BILAN' if 'BILAN' in df_comp.columns else banque_col, ascending=False)

    rename_map = {'BILAN': 'Total Bilan', 'EMPLOI': 'Emplois', 'RESSOURCES': 'Ressources',
                  'FONDS_PROPRES': 'Fonds Propres', 'produit_net_bancaire': 'PNB',
                  'resultat_net': 'Résultat Net', 'effectif': 'Effectifs', 'agence': 'Agences',
                  banque_col: 'Banque'}
    df_display = df_comp.rename(columns={k: v for k, v in rename_map.items() if k in df_comp.columns})

    for c in ['Total Bilan', 'Emplois', 'Ressources', 'Fonds Propres', 'PNB', 'Résultat Net']:
        if c in df_display.columns:
            df_display[c] = df_display[c].apply(lambda x: f"{x:,.0f}")
    for c in ['Effectifs', 'Agences']:
        if c in df_display.columns:
            df_display[c] = df_display[c].apply(lambda x: f"{int(x):,}" if pd.notna(x) else '')

    styled = df_display.style.set_properties(**{
        'font-size': '11px', 'text-align': 'right', 'padding': '8px 12px',
        'border-bottom': '1px solid #e2e8f0'
    }).set_properties(subset=['Banque'], **{'text-align': 'left', 'font-weight': '600'})

    styled = styled.set_table_styles([
        {'selector': 'th', 'props': [
            ('background', 'linear-gradient(135deg, #0f172a, #0891b2)'),
            ('color', 'white'), ('font-weight', '700'), ('font-size', '11px'),
            ('padding', '10px 12px'), ('text-align', 'center'), ('letter-spacing', '0.02em')]},
        {'selector': 'tr:nth-child(even)', 'props': [('background-color', '#f1f5f9')]},
        {'selector': 'tr:hover', 'props': [('background-color', 'rgba(44,82,130,0.04)')]},
        {'selector': 'table', 'props': [
            ('border-collapse', 'collapse'), ('width', '100%'),
            ('box-shadow', '0 4px 20px rgba(26,41,64,0.08)'),
            ('border-radius', '10px'), ('overflow', 'hidden')]},
    ])
    styled = styled.hide(axis='index')
    display(styled)
```
